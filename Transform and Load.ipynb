# Normalizing Genres

```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
# Load data
df = pd.read_csv('Data/basics.csv')
df.head()
```

```python
# Create a col with a list of genres
df['genres_split'] = df['genres'].str.split(',')
df
```

```python
exploded_genres = df.explode('genres_split')
exploded_genres
```

```python
# Get unique genres from the 'genres_split' column.
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
#Create a new title_genres from the exploded df
title_genres = exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
# Making the genre mapper dictionary
genre_ints=range(len(unique_genres))
genre_map=dict(zip(unique_genres,genre_ints))
genre_map
```

```python
# Make new interger genre_id and drop string genres
exploded_genres['genre_id'] = exploded_genres['genres_split'].replace(genre_map)
exploded_genres=exploded_genres.drop(columns='genres_split')
exploded_genres
```

```python
genre_lookup = pd.DataFrame({'genre_name':genre_map.keys(),
                             'genre_id':genre_map.values()})
genre_lookup.head()
```

```python
# Load API result data 
df_tmdb = pd.read_csv('Data/tmdb_results_combined.csv.gz')
df_tmdb.head()
```

```python
# Select columns to insert into sql database
tmdb = df_tmdb[['imdb_id','budget','revenue','certification']].copy()
tmdb
```

```python
tmdb.drop(0, inplace=True)
```

```python
tmdb.drop_duplicates(subset=['imdb_id'], inplace=True)
```

```python
# Use 1 larger than max string column
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = tmdb['imdb_id'].fillna('').map(len).max()
cert_len = tmdb['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "imdb_id": CHAR(key_len+1), 
    "budget": FLOAT(),
    "revenue": FLOAT(),
    'certification':CHAR(key_len+1)}
```

```python
# Save to sql with dtype and index=False
tmdb.to_sql('tmdb_data',conn, dtype=df_schema, if_exists='replace',index=False)
```

```python
# update the table and set the desired column as the primary key using sqlalchemy engine
conn.execute('ALTER TABLE tmdb_data ADD PRIMARY KEY (`imdb_id`);')
```
